```python
import matplotlib.pyplot as plt


def plot_pi_digits(start_point, end_point, plot_type='line'):
    """
    Plots the digits of Pi as a line graph or scatter plot.

    Parameters:
    start_point (int): The starting position of the digits.
    end_point (int): The ending position of the digits.
    plot_type (str): The type of plot ('line' or 'scatter').
    """
    pi_digits = pi.com_pi_digits(start_point, end_point)
    
    plt.figure(figsize=(10, 6))
    if plot_type == 'line':
        plt.plot(pi_digits, marker='o')
        plt.title(f"Line Graph of Pi Digits from {start_point} to {end_point}")
    elif plot_type == 'scatter':
        plt.scatter(range(len(pi_digits)), pi_digits, c='blue')
        plt.title(f"Scatter Plot of Pi Digits from {start_point} to {end_point}")
    
    plt.xlabel("Position in Pi")
    plt.ylabel("Digit Value")
    plt.grid(True)
    plt.show()
```

```python
plot_pi_digits(1,1000)
```

```python
plot_pi_digits(1,100)
```
